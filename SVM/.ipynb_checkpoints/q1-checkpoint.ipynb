To derive the maximum margin hyperplane, we need to find the distance of a point $x_i$ from the boundary $w^Tx+b=0$. This can be calculated using the formula $\left|\frac{w^Tx_i+b}{||w||}\right|$.

There are two ways to maximize the margin:

- Set the margin value to 1 and maximize the value of $\frac{1}{||w||}$, which is equivalent to minimizing $\frac{1}{2} w^T w$.
- Set the value of $||w||$ to 1 and maximize $\sum_i |w^Tx_i+b|$.

We will use the first approach, where we minimize the value of $\frac{1}{2} w^T w$ subject to the constraints that all points satisfy $y_i(w^Tx_i+b)\geq 1$ for all $i$, where $y_i$ represents the true class of the point $x_i$. These constraints enforce that all points are correctly classified and lie outside the margin.

By solving this optimization problem, we obtain the maximum margin hyperplane. This hyperplane is defined by the set of support vectors, which are the data points that lie closest to the hyperplane and define the margin. Any points that are not support vectors do not affect the hyperplane and lie outside the margin.

```python
# Write your code here
def solver(X, y):
  I = np.eye(X.shape[1])
  w = cp.Variable(X.shape[1])
  b = cp.Variable(1)

  cstrnt =  []
  for i in range(X.shape[0]):
    cstrnt.append(y[i] * (X[i]@w + b) >= 1)

  conds = cp.Minimize((1/2)*cp.quad_form(w, I))

  p = cp.Problem(conds,cstrnt)
  p.solve()
  return w.value, b.value

w, b = solver(X, y)
print("w: {}, b: {}".format(w, b))
```

In the next cell you are required to visualize the boundary predicted by the solver. You may want to revise up on how to plot the a line given w, b. Please plot w vector as well as the margin lines.

```python
# Write your code here
def line_of_sep_plot(X, y, w, b):
  x1 = np.linspace(X[:,0].min(), X[:,0].max())
  fact = (-w[0] / w[1]) * x1
  plt.plot(x1,fact - (b) / w[1],'orange')
  plt.plot(x1,fact - (b+1) / w[1],'b--')
  plt.plot(x1,fact - (b-1) / w[1],'r--')
  plot_points(X,y)
  plt.show()

line_of_sep_plot(X, y, w, b)
```

In the next cell you are required to modify your framing of optimisation problem to incorporate soft constraints also known as slack. Plot the decision boundary, w and margin lines for various values of C: $0.01$, $0.1$, $3$, $10$, $100$.

Explain your results theoretically. How does the boundary change with varying values of C? Why?

**Note: Use the modified X and y i.e after adding points in previous cell**

```python
def solver_with_soft_constraint(X, y, c):
  I = np.eye(X.shape[1])
  O = np.ones(X.shape[0])
  e = cp.Variable(X.shape[0])
  w = cp.Variable(X.shape[1])
  b = cp.Variable(1)

  cstrnt =  []
  err_cstrnt = []
  for i in range(X.shape[0]):
    cstrnt.append(y[i] * (X[i]@w + b) >= 1-e[i])
    err_cstrnt.append(e[i] >=0)

  cstrnt = cstrnt + err_cstrnt  
  conds = cp.Minimize((1/2)*cp.quad_form(w, I) + c * (e @ O))

  p = cp.Problem(conds,cstrnt)
  p.solve()
  return w.value, b.value
```

```python
# Write your code here for C= 0.01
w, b = solver_with_soft_constraint(X_new,y_new,0.01)
line_of_sep_plot(X_new, y_new, w, b)
```

```python
# Write your code here for C= 0.1
w, b = solver_with_soft_constraint(X_new,y_new,0.1)
line_of_sep_plot(X_new, y_new, w, b)
```

```python
# Write your code here for C= 3
w, b = solver_with_soft_constraint(X_new,y_new,3)
line_of_sep_plot(X_new, y_new, w, b)
```

```python
# Write your code here for C= 10
w, b = solver_with_soft_constraint(X_new,y_new,10)
line_of_sep_plot(X_new, y_new, w, b)
```

```python
# Write your code here for C= 100
w, b = solver_with_soft_constraint(X_new,y_new,100)
line_of_sep_plot(X_new, y_new, w, b)
```

When the value of C is high, errors are penalized more severely. This results in fewer points being permitted within the margin of the class. Conversely, lower values of C allow for a larger and softer margin, which permits more mistakes in the quest to expand the margin size.

Take the case of C = 100 and remove the support vectors. Does the boundary change after the deletion? Explain why?

```python
# Write your Code here

indexes_to_remove = [2, 7, 8]
keep_indexes = np.ones(X_new.shape[0], dtype=bool)
keep_indexes[indexes_to_remove] = False

X_new1 = X_new[keep_indexes]
y_new1 = y_new[keep_indexes]

w, b = solver_with_soft_constraint(X_new1,y_new1,100)
line_of_sep_plot(X_new1, y_new1, w, b)
```

Removing any of the support vectors would result in a change in the margin. Since the margin is just supported by support vectors, a new boundary appears upon deleting them.

```python
# Import Relevant Packages
from sklearn.svm import SVC
import numpy as np
from mnist.loader import MNIST
```

```python
mndata = MNIST('samples')

# Load training dataset
images, labels = mndata.load_training()
l = len(labels)
images_train = np.array(images)
labels_train = np.array(labels)

print(images_train.shape)
print(labels_train.shape)
```

```python
# Load testing dataset
images, labels = mndata.load_testing()
l = len(labels)
images_test = np.array(images)
labels_test = np.array(labels)

print(images_test.shape)
print(labels_test.shape)
```

- Tweak different parameters like the C Parameter and gamma parameter of the Linear SVM and report the results.

Report accuracy score, F1-score, Confusion matrix and any other metrics you feel useful.

```python
# use this subset of data for training and testing

batch = 3000
X_train = images_train[:batch,:]
y_train = labels_train[:batch]

X_test = images_test[batch:2*batch,:]
y_test = labels_test[batch:2*batch]
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix,r2_score
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
scaler = StandardScaler()
scaler.fit(np.vstack([images_train,images_test]))

def calc(C, kernel, gamma):
  print("Kernel: {}, Gamma: {}, C: {}".format(kernel, gamma, C))

  svm = SVC(C = C, kernel=kernel, gamma=gamma)
  svm.fit(scaler.transform(images_train[:6000]),labels_train[:6000])

  prediction = svm.predict(scaler.transform(images_test[:1000]))

  print('F1 Score:',f1_score(labels_test[:1000], prediction, average='micro'))
  print('Accuracy:',accuracy_score(labels_test[:1000], prediction))
  print('R squared:',r2_score(labels_test[:1000], prediction))
  print('Confusion Matrix:')
  sns.heatmap(confusion_matrix(labels_test[:1000], prediction))
  plt.show()
```

```python
for kernel in ['rbf', 'linear']:
  for gamma in [1e-3, 1e-5]:
      for C in [0.01,1,10]:
        calc(C, kernel, gamma)
```
